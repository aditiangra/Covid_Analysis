```python
df2['Prev_Confirmed'] = df2.groupby('State')['Confirmed'].shift(1)
df2['New Cases'] = df2['Confirmed'] - df2['Prev_Confirmed']
df2.drop('Prev_Confirmed',inplace = True,axis=1)
```

```python
df2['Prev_Recovered'] = df2.groupby('State')['Cured'].shift(1)
df2['New Recovered'] = df2['Cured'] - df2['Prev_Recovered']
df2.drop('Prev_Recovered',inplace = True,axis=1)
df2['Prev_Deaths'] = df2.groupby('State')['Deaths'].shift(1)
df2['New Deaths'] = df2['Deaths'] - df2['Prev_Deaths']
df2.drop('Prev_Deaths',inplace = True,axis=1)
```
